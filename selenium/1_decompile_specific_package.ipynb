```python
package_id = "0xa8ff9a23c6cfb36fb141585a5843be000e40a5424a868ce04aa42d5a25a11a70"
from selenium import webdriver
from selenium.webdriver.common.by import By
from time import sleep

driver = webdriver.Chrome()

driver.get(f"http://localhost:3000/decompile/{package_id}?api_key=suigpt_internal_testing")
```

```python
import time
import json
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By

# 初始化起始时间
start_time = time.time()

driver.find_element(By.ID, "suigpt_decompiled").click()
time.sleep(0.5)

print(f"Initial click took {time.time() - start_time:.2f} seconds.")

# 点击所有 module_button，然后点击 start_decompile，遍历所有 module_button
module_buttons = driver.find_elements(By.CLASS_NAME, "module_button")
for module_button in module_buttons:
    module_button.click()
    time.sleep(0.5)
    driver.find_element(By.ID, "start_decompile").click()

print(f"Decompilation started, took {time.time() - start_time:.2f} seconds.")

# 等待 ringAnimation_decompiler 消失
module_buttons = driver.find_elements(By.CLASS_NAME, "module_button")
for module_button in module_buttons:
    module_button.click()
    time.sleep(0.5)
    WebDriverWait(driver, 300).until_not(
        EC.presence_of_element_located((By.ID, "ringAnimation_decompiler"))
    )

print(f"Decompilation completed, took {time.time() - start_time:.2f} seconds.")

# 记录开始保存代码的时间
save_start_time = time.time()

# 保存所有模块的反编译代码为 JSON
decompiled_code_dict = {}
module_buttons = driver.find_elements(By.CLASS_NAME, "module_button")

for module_button in module_buttons:
    module_button.click()
    time.sleep(0.5)
    decompiled_code = driver.find_element(By.ID, "decompiled_code_area").text
    decompiled_code_dict[module_button.text] = decompiled_code

with open('decompiled_code.json', 'w') as json_file:
    json.dump(decompiled_code_dict, json_file, indent=4)

print(f"Decompiled code saved to decompiled_code.json, took {time.time() - save_start_time:.2f} seconds for saving.")
print(f"Total execution time: {time.time() - start_time:.2f} seconds.")
```
